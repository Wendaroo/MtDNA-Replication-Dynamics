- If you wish to simulate the posterior predictive trajectories manually from our saved posterior distribution sample (of any model), uncomment the model you wish to simulate

- If you wish for the run time to be reduced, set num_samples to be smaller. If you wish for more posterior predictive samples, set num_samples to be larger. 

- Stochastic three population pulse will take longer than the other models as it is an inhomogeneous Gillespie

```python
num_samples = 100
### -------- PULSE PORTION MODELS ----------- ###

#simulator = "one population pulse"
#simulator = "preliminary two population pulse"
#simulator = "two population pulse"
simulator = "three population pulse"
#simulator = "stochastic three population pulse"
#simulator = "dispersed three population pulse"

### -------- PULSE-CHASE PORTION MODELS ----------- ###

#simulator = "two population chase"
#simulator = "variable p two population chase"
#simulator = "three population chase"
```

```python
mode = simulator[-5:]
model = simulator[:-6]
```

Run the posterior predictive simulations. For num_samples = 100, this may take up to 5 minutes.

```python
if mode == "pulse": 
    (nucleoid_trajectories, edu_trajectories, peak1, variance_statistics) = ABC.posterior_predictive(num_samples = num_samples, model = model, inference_portion = "pulse", mode = "full")
elif mode == "chase":
    (nucleoid_trajectories, edu_trajectories, edu_proportions, peak1_0day, peak1_4day, variance_statistics) = ABC.posterior_predictive(num_samples = num_samples, model = model, inference_portion = "chase", mode = "full")
```

Plot the posterior predictive simulations

```python
if mode == "pulse":
    ABC_plot.single_posterior_predictive_pulse(nucleoid_trajectories, edu_trajectories, peak1, variance_statistics)
elif mode == "chase":
    ABC_plot.single_posterior_predictive_chase(nucleoid_trajectories, edu_trajectories, edu_proportions, peak1_0day, peak1_4day, variance_statistics)
```
