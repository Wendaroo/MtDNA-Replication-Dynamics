```python
import numpy as np
from sklearn.mixture import GaussianMixture
import matplotlib.pyplot as plt

from data_preprocessing.data_preprocessing import (intensities_0dy1, intensities_1dy1, intensities_2dy1, intensities_4dy1, intensities_0dy2
, intensities_1dy2,intensities_2dy2,intensities_4dy2,intensities_0dy3,intensities_1dy3,intensities_2dy3,intensities_4dy3)

def standard_normal_pdf(x):
    return 1/np.sqrt(2*np.pi) * np.exp(-x**2/2)

def normal_pdf(mu, sig, x):
    return standard_normal_pdf(np.divide(x-mu,sig))/sig

#Removing one anomalous point from 4 day assay 1 (this point is 2x larger than the next largest intensity)
anomaly = max(intensities_4dy1)
intensities_4dy1 = np.delete(intensities_4dy1, 4597)
new_max = max(intensities_4dy1)

if __name__ == "__main__":
    print("Anomalous intensity = " + str(anomaly))
    print("New maximum Intensity = " + str(new_max))

#Fitting Models
gm1_0dy = GaussianMixture(n_components=3, random_state=1).fit(intensities_0dy1.reshape((len(intensities_0dy1), 1)))
gm2_0dy = GaussianMixture(n_components=3, random_state=1).fit(intensities_0dy2.reshape((len(intensities_0dy2), 1)))
gm3_0dy = GaussianMixture(n_components=3, random_state=1).fit(intensities_0dy3.reshape((len(intensities_0dy3), 1)))

gm1_1dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_1dy1.reshape((len(intensities_1dy1), 1)))
gm2_1dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_1dy2.reshape((len(intensities_1dy2), 1)))
gm3_1dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_1dy3.reshape((len(intensities_1dy3), 1)))

gm1_2dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_2dy1.reshape((len(intensities_2dy1), 1)))
gm2_2dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_2dy2.reshape((len(intensities_2dy2), 1)))
gm3_2dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_2dy3.reshape((len(intensities_2dy3), 1)))

gm1_4dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_4dy1.reshape((len(intensities_4dy1), 1)))
gm2_4dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_4dy2.reshape((len(intensities_4dy2), 1)))
gm3_4dy = GaussianMixture(n_components=2, random_state=1).fit(intensities_4dy3.reshape((len(intensities_4dy3), 1)))

#Extracting weights and means
gm1_0dy_weights_ordered = gm1_0dy.weights_[np.argsort(np.transpose(gm1_0dy.means_)[0])]
gm1_0dy_means_ordered = np.transpose(gm1_0dy.means_)[0][np.argsort(np.transpose(gm1_0dy.means_)[0])]
gm2_0dy_weights_ordered = gm2_0dy.weights_[np.argsort(np.transpose(gm2_0dy.means_)[0])]
gm2_0dy_means_ordered = np.transpose(gm2_0dy.means_)[0][np.argsort(np.transpose(gm2_0dy.means_)[0])]
gm3_0dy_weights_ordered = gm3_0dy.weights_[np.argsort(np.transpose(gm3_0dy.means_)[0])]
gm3_0dy_means_ordered = np.transpose(gm3_0dy.means_)[0][np.argsort(np.transpose(gm3_0dy.means_)[0])]

gm1_1dy_weights_ordered = gm1_1dy.weights_[np.argsort(np.transpose(gm1_1dy.means_)[0])]
gm1_1dy_means_ordered = np.transpose(gm1_1dy.means_)[0][np.argsort(np.transpose(gm1_1dy.means_)[0])]
gm2_1dy_weights_ordered = gm2_1dy.weights_[np.argsort(np.transpose(gm2_1dy.means_)[0])]
gm2_1dy_means_ordered = np.transpose(gm2_1dy.means_)[0][np.argsort(np.transpose(gm2_1dy.means_)[0])]
gm3_1dy_weights_ordered = gm3_1dy.weights_[np.argsort(np.transpose(gm3_1dy.means_)[0])]
gm3_1dy_means_ordered = np.transpose(gm3_1dy.means_)[0][np.argsort(np.transpose(gm3_1dy.means_)[0])]

gm1_2dy_weights_ordered = gm1_2dy.weights_[np.argsort(np.transpose(gm1_2dy.means_)[0])]
gm1_2dy_means_ordered = np.transpose(gm1_2dy.means_)[0][np.argsort(np.transpose(gm1_2dy.means_)[0])]
gm2_2dy_weights_ordered = gm2_2dy.weights_[np.argsort(np.transpose(gm2_2dy.means_)[0])]
gm2_2dy_means_ordered = np.transpose(gm2_2dy.means_)[0][np.argsort(np.transpose(gm2_2dy.means_)[0])]
gm3_2dy_weights_ordered = gm3_2dy.weights_[np.argsort(np.transpose(gm3_2dy.means_)[0])]
gm3_2dy_means_ordered = np.transpose(gm3_2dy.means_)[0][np.argsort(np.transpose(gm3_2dy.means_)[0])]

gm1_4dy_weights_ordered = gm1_4dy.weights_[np.argsort(np.transpose(gm1_4dy.means_)[0])]
gm1_4dy_means_ordered = np.transpose(gm1_4dy.means_)[0][np.argsort(np.transpose(gm1_4dy.means_)[0])]
gm2_4dy_weights_ordered = gm2_4dy.weights_[np.argsort(np.transpose(gm2_4dy.means_)[0])]
gm2_4dy_means_ordered = np.transpose(gm2_4dy.means_)[0][np.argsort(np.transpose(gm2_4dy.means_)[0])]
gm3_4dy_weights_ordered = gm3_4dy.weights_[np.argsort(np.transpose(gm3_4dy.means_)[0])]
gm3_4dy_means_ordered = np.transpose(gm3_4dy.means_)[0][np.argsort(np.transpose(gm3_4dy.means_)[0])]

#Contructing some arrays with this information
means_0dy = np.array([gm1_0dy_means_ordered, gm2_0dy_means_ordered, gm3_0dy_means_ordered])
means_1dy = np.array([gm1_1dy_means_ordered, gm2_1dy_means_ordered, gm3_1dy_means_ordered])
means_2dy = np.array([gm1_2dy_means_ordered, gm2_2dy_means_ordered, gm3_2dy_means_ordered])
means_4dy = np.array([gm1_4dy_means_ordered, gm2_4dy_means_ordered, gm3_4dy_means_ordered])

weights_0dy = np.array([gm1_0dy_weights_ordered, gm2_0dy_weights_ordered, gm3_0dy_weights_ordered])
weights_1dy = np.array([gm1_1dy_weights_ordered, gm2_1dy_weights_ordered, gm3_1dy_weights_ordered])
weights_2dy = np.array([gm1_2dy_weights_ordered, gm2_2dy_weights_ordered, gm3_2dy_weights_ordered])
weights_4dy = np.array([gm1_4dy_weights_ordered, gm2_4dy_weights_ordered, gm3_4dy_weights_ordered])

peak1_mean_assay1 = np.array([gm1_0dy_means_ordered[0], gm1_1dy_means_ordered[0], gm1_2dy_means_ordered[0], gm1_4dy_means_ordered[0]])
peak1_mean_assay2 = np.array([gm2_0dy_means_ordered[0], gm2_1dy_means_ordered[0], gm2_2dy_means_ordered[0], gm2_4dy_means_ordered[0]])
peak1_mean_assay3 = np.array([gm3_0dy_means_ordered[0], gm3_1dy_means_ordered[0], gm3_2dy_means_ordered[0], gm3_4dy_means_ordered[0]])

peak2_mean_assay1 = np.array([gm1_0dy_means_ordered[1], gm1_1dy_means_ordered[1], gm1_2dy_means_ordered[1], gm1_4dy_means_ordered[1]])
peak2_mean_assay2 = np.array([gm2_0dy_means_ordered[1], gm2_1dy_means_ordered[1], gm2_2dy_means_ordered[1], gm2_4dy_means_ordered[1]])
peak2_mean_assay3 = np.array([gm3_0dy_means_ordered[1], gm3_1dy_means_ordered[1], gm3_2dy_means_ordered[1], gm3_4dy_means_ordered[1]])

peak1_weight_assay1 = np.array([gm1_0dy_weights_ordered[0], gm1_1dy_weights_ordered[0], gm1_2dy_weights_ordered[0], gm1_4dy_weights_ordered[0]])
peak1_weight_assay2 = np.array([gm2_0dy_weights_ordered[0], gm2_1dy_weights_ordered[0], gm2_2dy_weights_ordered[0], gm2_4dy_weights_ordered[0]])
peak1_weight_assay3 = np.array([gm3_0dy_weights_ordered[0], gm3_1dy_weights_ordered[0], gm3_2dy_weights_ordered[0], gm3_4dy_weights_ordered[0]])

all_peak1_means = np.concatenate([peak1_mean_assay1, peak1_mean_assay2, peak1_mean_assay3])
all_peak2_means = np.concatenate([peak2_mean_assay1, peak2_mean_assay2, peak2_mean_assay3])

#The Relevant Information for the ABC summary statistics:
mean_0dy_peak1_prop = np.mean(weights_0dy[:,0])
var_0dy_peak1_prop = np.var(weights_0dy[:,0])

mean_4dy_peak1_prop = np.mean(weights_4dy[:,0])
var_4dy_peak1_prop = np.var(weights_4dy[:,0])

if __name__ == "__main__":
    print("--------------------------------------")
    print("0 day peak 1 proportions = " + str(weights_0dy[:,0]))
    print("0 day peak 1 proportion mean = " + str(mean_0dy_peak1_prop))
    print("0 day peak 1 proportion variance = " + str(var_0dy_peak1_prop))
    print("0 day peak 1 proportion mean err = " + str(np.sqrt(var_0dy_peak1_prop/3)))
    print("--------------------------------------")
    print("4 day peak 1 proportions = " + str(weights_4dy[:,0]))
    print("4 day peak 1 proportion mean = " + str(mean_4dy_peak1_prop))
    print("4 day peak 1 proportion variance = " + str(var_4dy_peak1_prop))
```
